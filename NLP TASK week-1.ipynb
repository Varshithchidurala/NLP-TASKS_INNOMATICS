```python
1. Difference between "Love" and "love" in NLP

In NLP, "Love" and "love" are treated as two different words because computers see them as separate tokens.
For example, "Love" may appear at the beginning of a sentence, while "love" appears in the middle.
If we don’t convert everything to lowercase, the model may think they are different words even though they mean the same thing. 
That’s why we usually normalize text by converting it to lowercase.
```

```python
2.What happens if stopwords are not removed?

If stopwords are not removed, the text will contain a lot of unnecessary words like "the", "is", "and", etc.
These words don’t carry much meaning but appear very frequently, which can make the model focus on unimportant information.
This increases noise in the data and may reduce model performance, especially in tasks like text classification.
```

```python
3.Provide two real-world scenarios where removing stopwords can be harmful.

1. Sentiment Analysis:
In sentences like "I am not happy", if we remove the word "not", the sentence becomes "I am happy", which completely changes the meaning. So removing stopwords can lead to incorrect predictions.

2. Question Answering Systems:
In questions like "What is your name?", removing words like "what" or "is" makes the sentence unclear. These words are important to understand the intent of the question.
```

```python
4.Explain the difference between stemming and lemmatization.

Stemming and lemmatization are both used to reduce words to their base form, but they work differently.

Stemming is a simple and faster method that cuts words without caring about whether the result is a real word. For example, "running" may become "run" or sometimes even "runn".

Lemmatization is more advanced and uses language rules to convert a word into its proper base form. For example, "running" becomes "run", and "better" becomes "good".

So, stemming is faster but less accurate, while lemmatization is slower but more meaningful.
```

```python
#TASK-2

import re

def preprocess_text(text):

    if not text or not isinstance(text, str):
        return [], ""

    # lowercase
    text = text.lower()

    # remove urls & emails
    text = re.sub(r"http\S+|www\S+|@\S+", "", text)

    # remove numbers
    text = re.sub(r"\d+", "", text)

    # fix repeated characters
    text = re.sub(r"(.)\1+", r"\1", text)

    # remove special characters
    text = re.sub(r"[^a-z\s]", "", text)

    # remove extra spaces
    text = re.sub(r"\s+", " ", text).strip()

    # tokenization
    tokens = text.split()

    # remove short tokens except no, not
    tokens = [t for t in tokens if len(t) > 2 or t in ["no", "not"]]

    clean_sentence = " ".join(tokens)

    return tokens, clean_sentence
```

```python
#TASK-3

test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

for text in test_sentences:
    tokens, clean = preprocess_text(text)

    print("\nOriginal:", text)
    print("Tokens:", tokens)
    print("Clean Sentence:", clean)
```

```python
#TASK-4

total_tokens = len(tokens)
unique_tokens = len(set(tokens))

if total_tokens > 0:
    avg_len = sum(len(t) for t in tokens) / total_tokens
else:
    avg_len = 0

print("Total Tokens:", total_tokens)
print("Unique Tokens:", unique_tokens)
print("Average Length:", round(avg_len, 2))
```

```python
#TASK_5

from collections import Counter

all_tokens = []

for text in test_sentences:
    tokens, _ = preprocess_text(text)
    all_tokens.extend(tokens)

counter = Counter(all_tokens)
print("Top 10 words:")
print(counter.most_common(10))
print("\nLeast 5 words:")
print(counter.most_common()[:-6:-1])
```

```python
#TASK-6

def full_pipeline(text_list):

    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, clean = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(clean)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }


result = full_pipeline(test_sentences)
print(result)
```
